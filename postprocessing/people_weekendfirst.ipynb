```python
# Read the CSV file
df_user = pd.read_csv('/Users/naroacorettisanchez/Documents/GitHub/MultifunctionalVehicleFleets/includes/Demand/user_demand_cambridge_oct7_2019_week.csv')
```

```python
df_user['starttime'] = pd.to_datetime(df_user['starttime'])
```

```python
dfs = {}

for day in range(7,13 +1):
    day_df= df_user[df_user['starttime'].dt.day == day].copy()
    dfs[day] = day_df
```

```python
# Define the order of the days
order = [12, 13, 7, 8, 9, 10, 11]

# Create a new DataFrame by concatenating the separate DataFrames in the specified order
new_df = pd.concat([dfs[day] for day in order])

# Reset the index of the new DataFrame
new_df.reset_index(drop=True, inplace=True)
```
