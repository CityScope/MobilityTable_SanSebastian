```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

Fact = 1.0  # You can choose any Fact value here

cd_values = [1/50, 1/100, 1/200] #0.02 0.01  0.005
cw_values = [1, 2, 3]
d_values = np.array([50, 100, 200, 300])
w_values = np.array([1, 2, 3])
```

```python


# Create meshgrid for d and w
d_mesh, w_mesh = np.meshgrid(d_values, w_values)

# Calculate the equation values
equation_values = Fact * (-cd_values[0] * d_mesh + cw_values[0] * w_mesh)  # Using the first values of cd and cw

# Create contour plot
plt.figure(figsize=(10, 6))
plt.title(f'Contour Plot for Equation (Fact = {Fact}, cd = {cd_values[0]}, cw = {cw_values[0]})')
plt.xlabel('d')
plt.ylabel('w')
contour = plt.contourf(d_mesh, w_mesh, equation_values, levels=20, cmap='viridis')
plt.colorbar(contour)
plt.grid(True)
plt.show()

```
